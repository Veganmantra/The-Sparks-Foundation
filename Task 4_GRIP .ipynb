# Data Visualisation

```python
data['Year'].value_counts(dropna = False).sort_index()
```

```python
x_year = data['Year'].unique()
y_count_years = data['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'dark')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack over the years', fontsize=20)
plt.show()
```

Year 2014 showed highest rate of attacks.

```python
data['Casualities'] = data['Killed'] + data['Wounded']

data.info()
```

```python
#Plot year and region casulatity
plt.subplots(figsize=(16,8))
data.groupby(['Region', 'Year']).count()['Casualities'].plot()
plt.xticks(rotation=90)
plt.ylabel("Casualities")
plt.title('Casualities as per the region', fontsize=20)
```

```python
data['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color=['turquoise'])
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 most affected city",fontsize=20)
plt.show()
```

```python
sns.barplot(data['Group'].value_counts()[1:15].values,data['Group'].value_counts()[1:15].index,palette=('Blues_d'))
plt.xticks(rotation=90)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.title('Terrorist Groups with Highest Terror Attacks',fontsize=20)
plt.show()
```

```python
label1=data['AttackType'].value_counts()[:5].index
size=data['AttackType'].value_counts()[:5].values
explode = (0.1, 0, 0, 0,0) 
fig1, ax1 = plt.subplots()
ax1.pie(size, explode=explode, labels=label1, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Types of attacks',fontsize=20)
plt.show()
```

We can see that Bombing/explosion are seen to be the cause of 51.4% of the deaths in these attacks.

```python
plt.rcParams['figure.figsize']=18,7
terror_area=pd.crosstab(data.Year,data.Region)
terror_area.plot()
plt.title('Trend of terror attacks in each region',fontsize=20)
plt.show()
```
